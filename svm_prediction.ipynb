```python
#svm model training --handwritten dataset
from sklearn.svm import SVC
import numpy as np
```

```python
import pandas as pd
digits=pd.read_csv('./train.csv')
#change the type to numpy arr
digits=digits.to_numpy()

#separate the target arr and feature arr
digits_y=digits[:, 0]
digits_x=digits[:, 1:]
```

```python
#since using the 'rbf' kernel, scaling the features first
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()  
digits_x = scaler.fit_transform(digits_x)  #normalized the features matrix
```

```python
#split the dataset
split_index=int(digits_x.shape[0]*0.8)
train_x=digits_x[:split_index]
train_y=digits_y[:split_index]

test_x=digits_x[split_index:]
test_y=digits_y[split_index:]
train_x.shape, train_y.shape, test_x.shape, test_y.shape
```

```python
#training the first model used to identify the number '0'

#first processing the target arr 
y_model_1=np.where(train_y!=0, 1, 0)     #change the label
svm_model_0=SVC(C=10, gamma='scale', kernel='rbf')
svm_model_0.fit(train_x, y_model_1)     
```

```python
#training the second model used to identify the number '1'
y_model_2=np.where(train_y!=1, 0, 1)
svm_model_1=SVC(C=10, gamma='scale', kernel='rbf')
svm_model_1.fit(train_x, y_model_2)
```

```python
#the third model
y_model_3=np.where(train_y==2, 1, 0)
svm_model_2=SVC(C=10, gamma='scale', kernel='rbf')
svm_model_2.fit(train_x, y_model_3)
```

```python
#the fourth model
y_model_4=np.where(train_y==3, 1, 0)
svm_model_3=SVC(C=10, gamma='scale', kernel='rbf')
svm_model_3.fit(train_x, y_model_4)
```

```python
#the 5th model
y_model_5=np.where(train_y==4, 1, 0)
svm_model_4=SVC(C=10, gamma='scale', kernel='rbf')
svm_model_4.fit(train_x, y_model_5)
```

```python
#the 6th model
y_model_6=np.where(train_y==5, 1, 0)
svm_model_5=SVC(C=10, gamma='scale', kernel='rbf')
svm_model_5.fit(train_x, y_model_6)
```

```python
#the 7th model
y_model_7=np.where(train_y==6, 1, 0)
svm_model_6=SVC(C=10, gamma='scale', kernel='rbf')
svm_model_6.fit(train_x, y_model_7)
```

```python
#the 8th model
y_model_8=np.where(train_y==7, 1, 0)
svm_model_7=SVC(C=10, gamma='scale', kernel='rbf')
svm_model_7.fit(train_x, y_model_8)
```

```python
#the 9th model
y_model_9=np.where(train_y==8, 1, 0)
svm_model_8=SVC(C=10, gamma='scale', kernel='rbf')
svm_model_8.fit(train_x, y_model_9)
```

```python
#the 10th model
y_model_10=np.where(train_y==9, 1, 0)
svm_model_9=SVC(C=10, gamma='scale', kernel='rbf')
svm_model_9.fit(train_x, y_model_10)
```

```python
svm_ovo = SVC(kernel='rbf', decision_function_shape='ovo')  
svm_ovo.fit(train_x, train_y)
```

```python
#prediction on testing data
y_pred=svm_ovo.predict(test_x)
acc=np.mean(y_pred==test_y.astype(int))
print(f'the accuracy for the testing data is {acc}')
```
